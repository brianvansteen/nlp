```python
# cleaning the tweet corpus, with function modified from Kaggle notebook:
# https://www.kaggle.com/code/andreshg/nlp-glove-bert-tf-idf-lstm-explained#9.-NLP:-Dis
def cleanText(text):

    '''Make text lowercase, remove text in square brackets,
    remove links, remove punctuation, remove newlines
    and remove words containing numbers.'''

    text = str(text).lower() # convert all text to lower case
    text = re.sub('\[.*?\]', '', text) # remove square-bracketed text
    text = re.sub('https?://\S+|www\.\S+', '', text) # remove URLs
    text = re.sub('<.*?>+', '', text) # remove HTML tags < >
    text = re.sub('(\s)?@\w+', '', text) # remove all @ references, needed before removing punctuation next line
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # remove punctuation
    text = re.sub('\n', '', text) # remove new line characters
    text = re.sub('\w*\d\w*', '', text) # remove words that contain digits
    return text
```

```python
# define stopwords
stop_words = stopwords.words('english')
more_stopwords = ['u', 'im', 'c', 'amp'] # u / you, im / i am, c / see, amp from &amp used as HTML special characters
stop_words = stop_words + more_stopwords
```

#### Using Snowball Stemmer

```python
# using the Snowball Stemmer, also known as the Porter2 stemming algorithm, an improved version of the Porter Stemmer
stemmer = nltk.SnowballStemmer("english")
```

```python
def preprocessTweets(text):

    '''
    Clean the text
    Remove the stop words
    Stem the words using the Snowballstemmer imported above
    '''
    
    text = cleanText(text) # call 'cleanText' function to clean words
    text = ' '.join(word for word in text.split(' ') if word not in stop_words) # split tweet, remove stop words, re-join words
    text = ' '.join(stemmer.stem(word) for word in text.split(' ')) # split tweet, apply the Snowball Stemmer, re-join words
    return text
```

```python
# create new column of cleaned tweets
dfTweets['clean'] = dfTweets['content'].apply(preprocessTweets)
dfTweets.head()
```

## 7. Classification Approach

The topic modelling functions are used from the course lessons, modified and updated for this project.

Count Vectorizer is used to convert the collection of 'clean' tweets into a matrix of token counts.

Latent Dirichlet Allocation (LDA) is a model for identifying the 'clean' tweets into a defined number of topics, based on the words in the tweet.

### 7.1 Latent Dirichlet Allocation and Count Vectorizer

```python
# create the LDA model
def fitLDA(data, num_topics):
    lda = LDA(n_components = num_topics, n_jobs = -1, random_state = 42)
    lda.fit(data)
    return lda
```

```python
# turn the 'clean' tweets into vectors
def countVectorizer(tweets):
    count_vectorizer = CountVectorizer(max_features = 1500)
    data = count_vectorizer.fit_transform(tweets)
    return (count_vectorizer, data)
```

```python
# function to get most common words in the tweets, using the LDA model, the CountVectorizer, and a set number of top words
def get_most_common_words_for_topics(model, vectorizer, n_top_words):
    words = vectorizer.get_feature_names_out()
    word_dict = {}
    for topic_index, topic in enumerate(model.components_): # pseudo count for the number of times a word 'j' was assigned to topic 'i'
        this_topic_words = [words[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        word_dict[topic_index] = this_topic_words
    return word_dict

# print topics and words, based on output return from above function
def print_topic_words(word_dict):
    for key in word_dict.keys():
        print(f"Topic {key}")
        print()
        print("\t", word_dict[key])
        print()
```

```python
topics = 5 # categorize to five topics
```

```python
tweets = dfClean['clean'] # use the 'clean' column of tweets from the dataframe
```

```python
# call the countVectorizer function, returning the vectorizer model, and the data matrix
(vectorizer, data) = countVectorizer(tweets)
```

```python
# showing the 99,999 tweets, in a sparse matrix of 1,500 defined as 'max_features'
data
```

```python
# the vectorizer model
vectorizer
```

```python
# call the LDA function, using the parameters 'data' that is returned from the countVectorizer function, and 'topics' that is a variable defined above
lda = fitLDA(data, topics)
```

```python
# get the most common topic words, based on the LDA model and CountVectorizer
topic_words = get_most_common_words_for_topics(lda, vectorizer, 20)
print_topic_words(topic_words)
```

### 7.2 Latent Dirichlet Allocation and TF-IDF Vectorizer

As a next analysis, the Term Frequency - Inverse Document Frequency (TF-IDF) method will be used as a vectorizer instead of the Count Vectorizer. TF-IDF measures the importance of a term in a document, based on the frequency of the term in the document (TD) and the inverse document frequency (IDF) of the term across the entire corpus. This helps to identify which terms are most relevant to a particular document.

For this analysis, the 'trucker' word will be added to the 'stop words' list.

Additionally, stemming will not be used for this analysis.

```python
# define stopwords, including 'trucker'
stop_words = stopwords.words('english')
more_stopwords = ['u', 'im', 'c', 'amp', 'trucker', 'truckers'] # u / you, im / i am, c / see, amp from &amp used as HTML special characters
stop_words = stop_words + more_stopwords
```

```python
def preprocessTweets(text):
    text = cleanText(text) # call 'cleanText' function to clean words
    text = ' '.join(word for word in text.split(' ') if word not in stop_words) # split tweet, remove stop words, re-join words
    return text
```

```python
# re-confirm dataframe, to define new corpus based on the 'content' column
dfTweets.head()
```

```python
# create new column 'tfidf' of cleaned tweets, without stemming
dfTweets['tfidf'] = dfTweets['content'].apply(preprocessTweets)
dfTweets.head()
```

```python
# convert the tweets into a TF-IDF vector using the given n-gram range
def tfidfVectorizer(documents, ng_range=(1, 1)):
    tfidf_vectorizer = TfidfVectorizer(ngram_range = ng_range, smooth_idf = True)
    tfidf_data = tfidf_vectorizer.fit_transform(documents)
    return (tfidf_vectorizer, tfidf_data)
```

```python
# function to preprocess the text
def preprocessText(data):
    '''
    Tokenize the text
    Remove the stop words
    Stem the words using the Snowballstemmer imported above
    '''
    
    # tokenization the text
    tk = RegexpTokenizer('\s+', gaps = True)
    text_data = [] # list for storing the tokenized data
    for values in data.text:
        tokenized_data = tk.tokenize(values) # Tokenize the news
        text_data.append(tokenized_data) # append the tokenized data

    # Removing stopwords from the 'text_data' list
    sw = stopwords.words('english')
    clean_data = [] # List for storing the clean text
    for data in text_data:
        clean_text = [words.lower() for words in data if words.lower() not in sw]
        clean_data.append(clean_text) # Appned the clean_text in the clean_data list
  
    # stem the text from the 'clean_data' list, using Snowballstemmer from above, 'stemmer = nltk.SnowballStemmer("english")'
    stemmed_data = [] # List for storing the stemmed data
    for data in clean_data:
        stemmed_text = [stemmer.stem(words) for words in data] # Stem the words
        stemmed_data.append(stemmed_text) # Append the stemmed text

    return stemmed_data
```

```python
# merge the training and test data into new dataframe for pre-processing, excluding the 'label' column from trainText and the 'id' column from testText, and re-set index
mergedData = pd.concat((trainText.drop('label', axis=1), testText.drop('id', axis=1)), axis=0).reset_index().drop('index', axis=1)
```
